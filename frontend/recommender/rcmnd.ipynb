```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/product_sales.csv - Sheet1 (1).csv')
print(df.head())
```

```python
pivot_df = df.pivot_table(
    index='Order id',
    columns='Product',
    values='quantity',
    aggfunc='count'
)
pivot_df.head()
```

```python
pivot_df.reset_index(inplace=True)
pivot_df = pivot_df.fillna(0)
pivot_df = pivot_df.drop('Order id', axis=1)
```

```python
pivot_df.head(2)
```

```python
pivot_df.rename(columns = {" Sup ": "Sup"}, inplace = True)
```

```python
co_matrix = pivot_df.T.dot(pivot_df)
np.fill_diagonal(co_matrix.values, 0)
```

```python
cos_score_df = pd.DataFrame(cosine_similarity(co_matrix))
cos_score_df.index = co_matrix.index
cos_score_df.columns = np.array(co_matrix.index)
```

```python
product_recs = []
for i in cos_score_df.index:
    product_recs.append(cos_score_df[cos_score_df.index!=i][i].sort_values(ascending = False)[0:5].index)
    
product_recs_df = pd.DataFrame(product_recs)
#product_recs_df['recs_list'] = product_recs_df.values.tolist()
product_recs_df.index = cos_score_df.index
```

```python
np.array(product_recs_df[product_recs_df.index == "Sup"])
```

```python
product_recs_df.index = product_recs_df.index.str.strip()
```

```python
product_recs_df = product_recs_df.replace("Red Tokri \r", "Red Tokri")
```
